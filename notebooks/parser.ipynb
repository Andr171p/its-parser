```python
import requests
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://its.1c.ru"
```

```python
GARANT_URL = "https://its.1c.ru/db/garant"
```

```python
response = requests.get(GARANT_URL)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
links_soup = soup.find("ul", attrs={"class": "nav list"})

links = [BASE_URL + link["href"] for link in links_soup.find_all("a")]

links
```

```python
response = requests.get(links[0])

soup = BeautifulSoup(response.text, "html.parser")

ul_soup = soup.find("ul", attrs={"class": "nav list"})

chapter_links = [BASE_URL + link["href"] for link in ul_soup.find_all("a")]

chapter_links
```

```python
response = requests.get(chapter_links[0])

soup = BeautifulSoup(response.text, "html.parser")

docs_soup = soup.find("ul", attrs={"class": "nav list"})

docs_links = [BASE_URL + link["href"] for link in docs_soup.find_all("a")]

docs_links
```

```python
from langchain_community.document_loaders import WebBaseLoader


loader = WebBaseLoader(docs_links[0])

doc = loader.load()

doc
```
